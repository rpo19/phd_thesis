```python
from gatenlp import Document, AnnotationSet

# --- 1) Plain text to annotate (NO LaTeX macros here!) ---
txt = """Giovanni Bianchi, residing in Bologna [...] -- plaintiff
and
Davide Ricci, residing in Milan [...] -- defendant
FACTS
The plaintiff summoned the defendant to court, alleging that on April 3, 2022 he lent him the sum of EUR 10,000, as evidenced by a private written agreement signed by both parties, with the obligation to repay by December 31, 2022, pursuant to Article 1813 c.c. [...]
"""

# --- 2) Create document ---
doc = Document(txt)
annset = doc.annset()              # default annotation set

# --- 3) Helper to add an annotation by substring match once --
def add_by_span(label, string, **features):
    idx = txt.index(string)
    annset.add(idx, idx+len(string), label, features)

# --- 4) Add annotations (one per span) ---
add_by_span("Person",   "Giovanni Bianchi",
            link=None)
add_by_span("Location", "Bologna",
            link="https://en.wikipedia.org/wiki/Bologna")

add_by_span("Person",   "Davide Ricci",
            link=None)
add_by_span("Location", "Milan",
            link="https://en.wikipedia.org/wiki/Modena")  # from your example

add_by_span("Date",     "April 3, 2022")
add_by_span("Money",    "EUR 10,000", currency="EUR")
add_by_span("Date",     "December 31, 2022")
add_by_span("Law",      "Article 1813 c.c.",
            source="https://www.normattiva.it/...")

# --- 5) Now the doc contains annotations ---
print(doc)               # debug
for a in annset:
    print(a)
```
